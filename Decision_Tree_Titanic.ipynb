```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
```

```python
df.head()
```

```python
inputs = df.drop('Survived',axis='columns')
target = df.Survived
```

```python
inputs.Sex = inputs.Sex.map({'male': 1, 'female': 2})
```

```python
inputs.Age[:10]
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.2)
```

```python
# dividing the datasets into two parts i.e. training datasets and test datasets
X, y = df, df['Survived']


# Splitting arrays or matrices into random train and test subsets
from sklearn.model_selection import train_test_split
# i.e. 70 % training dataset and 30 % test datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30)
```

```python
# importing random forest classifier from assemble module
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
```

```python
# prompt: tell me code to create a dataframe for the titanic dataset with Pclass, Sex, Age, Fare as inputs and Survived as target

inputs = df[['Pclass', 'Sex', 'Age', 'Fare']]
target = df['Survived']
```

```python
print(inputs.head())
print(target.head())
```

```python
# creating a RF classifier
clf = RandomForestClassifier(n_estimators = 100)

# Convert 'Sex' column to numerical using one-hot encoding
X_train = pd.get_dummies(X_train, columns=['Sex'], drop_first=True)
X_test = pd.get_dummies(X_test, columns=['Sex'], drop_first=True)

# Training the model on the training dataset
# fit function is used to train the model using the training sets as parameters
clf.fit(X_train, y_train)

# performing predictions on the test dataset
y_pred = clf.predict(X_test)

# metrics are used to find accuracy or error
from sklearn import metrics
print()

# using metrics module for accuracy calculation
print("ACCURACY OF THE MODEL:", metrics.accuracy_score(y_test, y_pred))
```
